   This grouping information appears immediately below, having been removed

   of size 369, while Group 1 has only 367 instances.  This is because it

   statements summarizes changes to the original Group 1's set of data:

   * Group 1 : 367 points: 200B 167M (January 1989)

```python
import pandas as pd
import numpy as np

bcw_df = pd.read_csv('breast-cancer-wisconsin.data', header=None)
bcw_df.info()
```

```python
bcw_df.head()
# all are discrete variable
```

```python
col_names = ['id', 'clump_thickness', 'cell_size', 'cell_shape', 'MA', 'SE_cell_size', 
             'bare_nuclei', 'bland_chromatin', 'normal_nucleoli', 'mitoses', 'malignant']
bcw_df.columns = col_names

# no need id
bcw_df = bcw_df.drop('id', axis=1)

# handle missing values 
bcw_df.replace('?', np.nan, inplace=True)
bcw_df['bare_nuclei'] = bcw_df['bare_nuclei'].astype('float64')
print("Missing values of Bare Nuclei: {}".format(bcw_df.loc[bcw_df.bare_nuclei.isnull()].malignant.count()))
```

#### Bare nuclei (Naked Nucleus):
are typical of undifferentiated/small cell carcinomas of any site—e.g., in small cell carcinoma of lung and elsewhere—and usually have friable—‘taffy-pull’—nuclei and scant or absent cytoplasm

#### Imputation

As there's no way to be sure whether data really are missing at random, or whether the missingness depends on unobserved predictors or the missing data themselves.
In a paper* I found, four missing value replacement methods were used, and median imputation gave the best result of 99.06% (Percentage of Correct Classification)

* Paulin, F. and Santhakumaran, A. 2010. An Algorithm to Reconstruct the Missing Values for Diagnosing the Breast Cancer

```python
test_bare_nuclei = bcw_df.loc[~bcw_df.bare_nuclei.isnull()].bare_nuclei
print("median value for the Bare Nuclei: {}".format(test_bare_nuclei.median()))
```

```python
bcw_df.malignant = bcw_df.malignant.map({4: 1, 2: 0})
display(pd.DataFrame(bcw_df.malignant.value_counts()))
```

## In this case I decide to use supervised classsification learning, since we have a well labeled data here, also the data is small enough, a plain batch learning is fine.

```python
import warnings
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    import statsmodels.api as smt
    from sklearn.model_selection import cross_val_score
    warnings.filterwarnings("ignore",category=FutureWarning)
    from sklearn.linear_model import SGDClassifier
    
from sklearn.model_selection import train_test_split  
from sklearn import metrics
from sklearn import model_selection
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.preprocessing import Imputer 
```

```python
# try to fit a logistic regression, to see how well are these variables can explain the malignant or not

imputer = Imputer(strategy="median")
imputer.fit(bcw_df.loc[:, bcw_df.columns != 'malignant'])
X=imputer.transform(bcw_df[['clump_thickness', 'cell_size', 'cell_shape', 'MA', 'SE_cell_size', 'bare_nuclei',
                            'bland_chromatin', 'normal_nucleoli', 'mitoses']])
y=bcw_df['malignant']

    
logit_model=smt.Logit(y,X)
result=logit_model.fit()
print(result.summary())
```

```python
# get train and test data sets
# and use sklearn's logistic regression model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

logreg = LogisticRegression()
logreg.fit(X_train, y_train);
```

```python
kfold = model_selection.StratifiedKFold(n_splits=10, random_state=7)
results = model_selection.cross_val_score(logreg, X_train, y_train, cv=kfold, scoring='accuracy')
print("10-fold cross validation average accuracy: {:.3f}".format(results.mean()))
```

```python
y_pred = logreg.predict(X_test)
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
log_fpr, log_tpr, log_thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(log_fpr, log_tpr, label='Logistic Regression (area = {:.3f})'.format(logit_roc_auc))
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

this model is pretty powerful, but let's try some more complex models, see if we can get a more accurate classifier.

```python
print('Accuracy of logistic regression classifier on test set: {:.3f}'.format(logreg.score(X_test, y_test)))
```

# TensorFlow

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import LeakyReLU
from keras.optimizers import SGD
from sklearn.preprocessing import StandardScaler
```

```python
from keras.optimizers import RMSprop
model = Sequential()
model.add(Dense(units = 7, kernel_initializer = 'RandomNormal', activation = 'sigmoid', input_dim = 9))
model.add(Dense(units = 6, kernel_initializer = 'RandomNormal', activation = 'relu'))
model.add(Dense(units = 1, kernel_initializer = 'RandomNormal', activation = 'sigmoid'))
model.compile(optimizer = RMSprop(), loss = 'binary_crossentropy', metrics = ['accuracy'])

model.fit(X_train, y_train, validation_split=0.2, batch_size = 7, epochs=120, verbose=2)
```

See more about hyperparameters at "grid_batch_epochs.py"

```python
y_pred = model.predict(X_test)
y_pred = [ 1 if y>=0.5 else 0 for y in y_pred ]
print(confusion_matrix(y_test, y_pred))
```

```python
scores = model.evaluate(X_test, y_test, verbose=0)
print('Accuracy of Keras Classifier on test set: {:.3f}'.format(scores[1]))
```

```python
model_yaml = model.to_yaml()
with open("model.yaml", "w") as yaml_file:
    yaml_file.write(model_yaml)
model.save_weights("model.h5")
```

```python
from keras.models import model_from_yaml
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=2)

with open("model_2.yaml", "r") as yaml_file:
    loaded_model_yaml = yaml_file.read()
    model_2_ex = model_from_yaml(loaded_model_yaml)
model_2_ex.load_weights("model_2.h5")
model_2_ex.layers.pop()
model_2_ex.outputs = [model_2_ex.layers[-1].output]
model_2_ex.layers[-1].outbound_nodes = []
model_2_ex.add(Dense(units = 1, kernel_initializer = 'RandomNormal', activation = 'sigmoid'))
for layer in model_2_ex.layers[:2]:
    layer.trainable = False


model_2_ex.compile(optimizer = 'Nadam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model_2_ex.fit(X_train_2, y_train_2, validation_split=0.2, batch_size = 1, epochs=50,
               callbacks=[early_stopping], verbose=2)
```

```python
y_pred = model_2_ex.predict(X_test_2)
y_pred = [ 1 if y>=0.5 else 0 for y in y_pred ]
print(confusion_matrix(y_test_2, y_pred))
```

```python
scores = model_2_ex.evaluate(X_test_2, y_test_2, verbose=0)
print('Accuracy of Keras Classifier on test set: {:.3f}'.format(scores[1]))
```

```python
roc_auc = roc_auc_score(y_test_2, y_pred)
print(roc_auc)
print(classification_report(y_test_2, y_pred))
```
